```python
#Gerekli paketler çağırılır.
from urllib.parse import urlparse,urlencode
import ipaddress
import re
#re --> Düzenli ifadeler (Regular Expressions)tüm modern dillerde bulunur
#dağınık bir metin içerisinde istediğimiz formattaki metinleri yakalayabilmemize imkan tanır.
```

### Domain Of The URL
URL'de bulunan etki alanını çıkarıyoruz. Bu özelliğin eğitimde pek bir önemi yoktur. Modeli eğitirken bile düşebilir.

```python
def getDomain(url):
  domain = urlparse(url).netloc # netloc--> Network location part,isteğe bağlı kimlik bilgileri ile birlikte domain name (ve varsa subdomain), bağlantı noktası numarasını içerir.Netloc: Host ismi ve bağlanılan port
  if re.match(r"^www.",domain):  # ^İfadenin başlangıcını kontrol eder ve [ ] karakterleri ile birlikte kullanılırsa da hariç anlamına gelir.
    domain = domain.replace("www.","")
  return domain
```

### IP Address In The URL
URL'deki IP varlığı kontrol edilir.URLs alan adı yerine IP adresi içerebilir.

URL içerisinde Internet Protokol
(IP) adresi geçiyor ise kullanıcının bilgileri çalınma
ihtimali bulunmaktadır.

```python
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip
```

### "@" Symbol In The URL
URL'deki "@" varlığı kontrol edilir.

URL’de “@” sembolünün kullanılması, tarayıcının “@” sembolünden önceki her
şeyi görmezden gelmesine neden olur ve gerçek adres
genellikle “@” sembolünü takip eder.


```python
def haveAtSign(url):
  if "@" in url:
    at = 1
  else:
    at = 0
  return at
```

### Length Of URL
Kimlik avcıları ya da dolandırıcılar, adres çubuğundaki şüpheli kısmı gizlemek
için uzun URL kullanabilirler. Bu yüzden dolayı URL uzunluğu da kontrol edilmektedir

```python
def getLength(url):
  if len(url) < 54:
    length = 0
  else:
    length = 1
  return length
```

### Depth Of URL
URL'nin derinliğini hesaplar.

Bu özellik verilen URL'deki '/' baz alarak alt sayfaları hesaplar.

```python
def getDepth(url):
  s =urlparse(url).path.split('/')
  depth=0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

### Redirection "//" In The URL
URL'deki '//' varlığını kontrol eder.

URL yolu içinde “//” ifadesinin bulunması da, kullanıcının başka bir web
sitesine yönlendirileceği anlamına gelir.

```python
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else :
      return 0
  else:
    return 0
```

### Http/Https In Domain Name
URL'nin alan adındaki "http/https" varlığı kontrol edilir.

Kimlik avcıları kullanıcıları kandırmak için URL'nin alan adına "https" ekleyebilirler.

```python
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

### Using URL Shortening Services “TinyURL”
URL kısaltma, “World Wide Web” de, bir URL’in uzunluğunun önemli ölçüde
daha küçük hale getirilebileceği ve gerekli web sayfasına yönlendirebileceği bir yöntemdir. Böylelikle gerçek domain name gizlenmiş olur.

 Bu, kısa bir alan adı üzerinde, uzun URL'ye sahip web sayfasına bağlantı veren bir "HTTP Yönlendirmesi" aracılığıyla gerçekleştirilir.

```python
#URL kısaltma hizmetlerini tespit etmek için normal bir ifade modeli sağlıyoruz.
#Kısaltılmış URL'leri tanımlamak amacıyla yaygın URL kısaltma hizmetleriyle eşleşecek şekilde tasarlanmıştır.

shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
#r öneki Python'da ham dize değişmezini belirtir. Ters eğik çizgileri gerçek karakterler olarak ele almak için kullanılır.
#Bir URL bu kalıbın herhangi bir bölümüyle eşleşiyorsa, bu hizmetlerden biri kullanılarak oluşturulmuş kısaltılmış bir URL olması muhtemeldir.
```

```python
# URL'deki kısaltma servislerini(shortening service) kontrol ediyoruz.
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
```

### Prefix or Suffix "-" in Domain
URL'deki "-" varlığını kontrol edilir.

“-“ sembolü meşru URL’lerde nadiren kullanılır. Kimlik avcıları,
kullanıcıların meşru bir web sayfasıyla uğraştıklarını hissetmeleri için alt alan adına (-) ile ayrılmış önekler(prefix) veya sonekler(suffix) ekleme eğilimindedir.

Eğer URL “-“ sembolü içeriyorsa 1, içermiyorsa 0 değeri döner.


```python
def prefixSuffix(url):
  if '-' in urlparse(url).netloc:
    return 1
  else:
    return 0
```

### **Web Traffic**
Bu özellik, ziyaretçi sayısını ve ziyaret ettikleri sayfa sayısını belirleyerek web sitesinin popülerliğini ölçer. Ancak phishing web siteleri kısa bir süre yaşadığı için Alexa veri tabanı tarafından tanınmayabilirler (Alexa the Web Information Company, 1996). Veri setimizi incelediğimizde en kötü senaryolarda meşru web sitelerinin ilk 100.000'de yer aldığını gördük. Ayrıca alan adının trafiği yoksa veya Alexa veri tabanı tarafından tanınmıyorsa "Kimlik Avı" olarak sınıflandırılır.

domain< 100000 ise bu özelliğin değeri 1 (phishing), aksi takdirde 0 (legimate) olur.

```python
def web_traffic(url):
  try:
    url= urllib.parse.quote(url)
    rank = BeautifulSoup(urllib.request.urlopen("http://data.alexa.com/data?cli=10&dat=s&url=" + url).read(),"xml").find("REACH")['RANK']
    rank= int(rank)
  except TypeError:
    return 1
  if rank < 100000:
    return 1
  else :
    return 0
#Belirli bir URL için Alexa verilerini getirir.
#BeautifulSoup kullanarak XML olarak ayrıştırır ve "REACH" etiketinden Alexa Sıralamasını çıkarır.
```

### **Age Of The Domain**

Kimlik avı web sitelerinin çoğu kısa bir süre için yayında kalır. Bu proje için meşru alan adının minimum yaşı 12 ay olarak kabul edilmektedir.

Alan adı yaşı > 12 ay ise bu özelliğin değeri 1 (kimlik avı), aksi halde 0 (yasal) olur.


```python
def domainAge(domain_name):
  creation_date = domain_name.create_date
  expiration_date= domain_name.expiration_date
  #Tarihlerin string olarak temsil edilip edilmediğini kontrol edin ve bunları datetime nesnelerine dönüştürelim
  if(isinstance(creation_date,str) or isinstance(expiration_date,str)):
    try:
      creation_date = datetime.strptime(creation_date,'%Y-%m-%d')
      expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
    except:
      return 1
  if ((expiration_date is None) or (creation_date is None)):
    return 1
  elif ((type(expiration_date)is list)or (type(creation_date)is list)):
    return 1
  else:
    ageofdomain = abs((expiration_date-creation_date).days)  #Belirtilen sayının mutlak değerini döndürür.
    if((ageofdomain/30) < 6) :
      age = 1
    else:
      age = 0
  return age
```

### **End Period of Domain**
Kalan domain süresi, sona erme süresi ile geçerli saat arasındaki fark bulunarak hesaplanır. Meşru alan adı için düşünülen bitiş süresi bu proje için 6 ay veya daha azdır.

Alan adının bitiş süresi 6 aydan fazla ise bu özelliğin değeri 1 (kimlik avı), aksi halde 0 (yasal) olur.

```python
def domainEnd(domain_name):
  expiration_date =domain_name.expiration_date
  if isinstance(expiration_date,str):
    try:
      expiration_date = datetime.strptime(expiration_date,"%Y-%m-%d")
    except:
      return 1
  if (expiration_date is None):
    return 1
  elif(type(expiration_date)is list):
    return 1
  else:
    today = datetime.now()
    end = abs((expiration_date-today).days)
    if((end/30)<6):
      end = 0
    else:
      end=1
    return end
```

### **IFrame Yönlendirmesi**
IFrame, ek bir web sayfasını o anda gösterilen sayfada görüntülemek için kullanılan bir HTML etiketidir.

Kimlik avcıları "iframe" etiketini kullanabilir ve çerçeve kenarlıkları olmadan onu görünmez hale getirebilir. Bu bağlamda kimlik avcıları, tarayıcının görsel bir tanımlama yapmasına neden olan “frameBorder” özelliğini kullanıyor.

iframe boşsa veya yanıt bulunamıyorsa bu özelliğe atanan değer 1 (phishing) veya 0'dır (legimate).

```python
def iframe(response):
  if response == "":
    return 1
  else:
    if re.findall(r"[<iframe>|<frameBorder>]",response.text):
      return 0
    else:
      return 1
```

### **Durum Çubuğu Özelleştirmesi**
Kimlik avı yapanlar, kullanıcılara durum çubuğunda sahte bir URL göstermek için JavaScript kullanabilir. Bu özelliği çıkarmak için web sayfasının kaynak kodunu kazıp durum çubuğunda herhangi bir değişiklik yapıp yapmadığını kontrol etmeliyiz.

Yanıt boşsa veya fareyle üzerine gelindiğinde bu özelliğe atanan değer 1 (phishing) veya 0'dır (legal).

```python
def mouseOver(response):
  if response == "":
    return 1
  else :
    if re.findall("<script>.+onmouseover.+<script>",response.text):
     return 1
    else:
     return 0
```

### **Sağ Tıklamayı Devre Dışı Bırakma**
Kimlik avı yapanlar, kullanıcıların web sayfası kaynak kodunu görüntüleyememesi ve kaydedememesi için sağ tıklama işlevini devre dışı bırakmak için JavaScript kullanır. Bu özellik için web sayfasının kaynak kodunda “event.button==2” olayını arayacağız ve sağ tıklamanın devre dışı olup olmadığını kontrol edeceğiz.

Yanıt boşsa veya fareyle üzerine gelindiğinde bulunamıyorsa bu özelliğe atanan değer 1 (phishing) veya 0'dır (legitimate).

```python
def rightClick(response):
  if response == "0":
    return 1
  else:
    if re.findall(r"event.button ?== ?2",response.text):
      return 0
    else:
      return 1
```

### **Websitesi Yönlendirme**
Phishing web sitelerini legitimate olanlardan ayıran ince çizgi, bir web sitesinin kaç kez yönlendirildiğidir.

Veri kümemizde meşru web sitelerinin en fazla bir kez yönlendirildiğini görüyoruz. Phishing web siteleri en az 4 kez yönlendirilmiştir.

```python
def forwarding(response):
  if response =="":
    return 1
  else:
    if len(response.history) <= 2:
      return 0
    else:
      return 1
```

# **URL Özelliklerini Hesaplama**
Bir liste ve diğer işlevleri çağıran ve URL'nin tüm özelliklerini listede saklayan bir fonksiyon oluşturalım. Her URL'nin özelliklerini çıkaracağız ve bu listeye ekleyeceğiz

```python
#Özellik Çıkarma Fonksiyonu
def featureExtraction(url,label):


  features = []
  features.append(getDomain(url))
  features.append(havingIP(url))
  features.append(haveAtSign(url))
  features.append(getLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(tinyURL(url))
  features.append(prefixSuffix(url))

  dns = 0
  try:
    domain_name = whois.whois(urlparse(url).netloc)
  except:
    dns =1

  features.append(dns)
  features.append(web_traffic(url))
  features.append(1 if dns == 1 else domainAge(domain_name))
  features.append(1 if dns == 1 else domainEnd(domain_name))

  try:
    response = requests.get(url)
  except:
    response = ""

  features.append(iframe(response))
  features.append(mouseOver(response))
  features.append(rightClick(response))
  features.append(forwarding(response))
  features.append(label)

  return features
```

```python
legi_features = []
label = 0

for i in range(0,5000):
  url = legiurl['URLs'][i]
  legi_features.append(featureExtraction(url,label))
```

```python
#Listeyi dataframe'e çevirelim

feature_names = ['Domain', 'Have_IP', 'Have_At', 'URL_Length', 'URL_Depth','Redirection',
                      'https_Domain', 'TinyURL', 'Prefix/Suffix', 'DNS_Record', 'Web_Traffic',
                      'Domain_Age', 'Domain_End', 'iFrame', 'Mouse_Over','Right_Click', 'Web_Forwards', 'Label']

legitimate = pd.DataFrame(legi_features, columns= feature_names)
legitimate.head()
```
